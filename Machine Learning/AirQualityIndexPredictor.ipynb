```python
import pandas as pd
```

```python
df = pd.read_csv("city_day.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
data = df.copy()
data.head()
```

```python
data.describe()
```

```python
data['AQI_Bucket'].value_counts()
```

```python
data['AQI_Bucket'].mode()
```

```python
data = data.apply(lambda x:x.fillna(x.median())
                            if x.dtype == 'float'
                            else x.fillna(x.mode()[0]))
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
airAQI     = (data['AQI_Bucket'])
airAQISet  = set(airAQI)
airAQIList = list(airAQISet)
dic1       = {airAQIList[i]:i for i in range(len(airAQISet))}
data['AQI_Bucket'] = data['AQI_Bucket'].map(dic1)
```

```python
airCity     = (data['City'])
airCitySet  = set(airCity)
airCityList = list(airCitySet)
dic2       = {airCityList[i]:i for i in range(len(airCitySet))}
data['City'] = data['City'].map(dic2)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import sklearn
```

```python
indep = data[['City', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]
indep.head()
```

```python
dep = data[['AQI']]
dep.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep, dep, test_size=0.2, random_state=42, shuffle=True)
```

```python
train_data = pd.concat([x_train, y_train], axis='columns',
                names = ['City', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI_Bucket', 'AQI'])
```

```python
train_data.head()
```

```python
corr_matrix = train_data.corr(method = 'pearson')
corr_matrix
```

```python
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```

```python
plt.scatter(x_train['CO'], y_train)
plt.title('CO vs AQI')
plt.xlabel('CO')
plt.ylabel('AQI')
plt.plot()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train, y_train)
y_pred_train = model.predict(x_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse_train = mean_squared_error(y_train, y_pred_train)
mae_train = mean_absolute_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

# Print the evaluation metrics on train data
print("Evaluation metrics: Train Data")
print("Mean absolute error:", mae_train)
print("Mean squared error :", mse_train)
print("R-squared score    :", r2_train)
```

```python
mse_test = mean_squared_error(y_test, y_pred)
mae_test = mean_absolute_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)

# Print the evaluation metrics on test data
print("Evaluation metrics: Test Data")
print("Mean absolute error:", mae_test)
print("Mean squared error :", mse_test)
print("R-squared score    :", r2_test)
```
